```python
"""
# Problem - 01

Given a list of integers, extract all numbers that are divisible by 3 but not divisible by 5.

Input Format:
Space-separated integers

Output Format:
A list of numbers satisfying the condition

Example:
Input:
3 5 9 10 15 18 20 21

Output:
[3, 9, 18, 21]


"""
num = input('Please enter your desire number: ')
num_list = list(map(int, num.split()))

new_list = []
for num in num_list:
    if num % 3 == 0 and num % 5 != 0:
        new_list.append(num)

print(new_list)

```

```python
"""
# Problem - 02

Write a function compute(a, b) that:
Returns (a² + b²) / (a - b)
Use exception handling to manage:
Division by zero

Input Format:
Two values

Output Format:
Computed result OR error message

"""
try:
    inp = input('Please enter your desire number: ')
    num = list(map(int,inp.split()))

    def compute(a,b):
        return (a**2 + b**2)/ (a - b)

    compute_result = compute(num[0], num[1])
    print(compute_result)

except ZeroDivisionError:
    print('You can not divided by zero.')

except Exception as e:
    print(e)

```
