```python
""""
Problem 1 (Easy - Warmup)

👉 Palindrome check (Two Pointer)

Input:

level

👉 Output কী হবে?

🎯 তোমার কাজ:
Two pointer ব্যবহার করো
loop দিয়ে solve করো

💡 Hint:
left = 0
right = len(s) - 1
while left < right
"""

inp = input()

left = 0
right = len(inp)-1

print(inp)
print(right)
isPalindrome = True

while left < right:
    if inp[left].lower() != inp[right].lower():
        isPalindrome = False
        break

    left += 1
    right -= 1
    
if isPalindrome:
    print('Palindrome')

else:
    print('not a palindrome')
```
