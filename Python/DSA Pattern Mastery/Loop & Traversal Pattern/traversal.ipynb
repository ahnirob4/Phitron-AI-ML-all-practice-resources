### Reverse Traversal

```python
arr = [2, 5, 7, 1]

for i in range(len(arr)-1, -1, -1):  #last index: first index: reverse step
    print(arr[i])
```

```python
inp = input()
num = list(map(int,inp))

reverse = []
for i in range(len(num)-1, -1,-1):
    reverse.append(num[i])

print(num)
print(reverse)
if num == reverse:
    print('this is mirror number')

else:
    print('Not a mirror')
```

```python
# Problem - 01
# 👉 reverse print করো

arr = [1, 2, 3, 4, 5]

for i in range(len(arr)-1, -1,-1):
    print(arr[i])

# Probblem - 02
# 👉 last element print করো (loop use করে)

arr = [10,20,30]

for i in range(len(arr)): # reverse arr[1-i+1] 
    if i == len(arr)-1: # লাস্ট ইনডেক্স বের করতে হলে condition ব্যাবহার করতে হবে। 
        print(arr[i])

# Problem - 03
# 👉 reverse করে নতুন list বানাও
 
arr = [7, 8, 9, 10]

new_list = []

for i in range(len(arr)-1, -1, -1):
    new_list.append(arr[i])

print(new_list)

# Problem - 04
# 👉 check করো palindrome কিনা (hint: reverse compare)

arr = [1, 2, 3, 2, 1]

reverse = []

for i in range(len(arr) -1, -1, -1):
    reverse.append(arr[i])

if arr == reverse:
    print('Palindrome')
else:
    print('not a palindrome')

# Problem - 05
# 👉 reverse traversal করে max বের করো

arr = [5, 3, 8, 1]

max = arr[0]

for i in range(len(arr)-1, -1,-1):
    if arr[i] > max:
        max = arr[i]
    print(arr[i], end=' ')
print(f'\n{max}')
```

### Conditional Traversal

```python
# Problem - 01
# 👉 10 এর বেশি number print করো

arr = [5, 12, 7, 18, 3]

for i in arr:
    if i > 10:
        print(i)

# problem - 02
# 👉 odd number গুলো count করো

arr = [2, 3, 4, 5, 6]

count = 0
for i in range(len(arr)):
    if arr[i]%2 == 1:
        count += 1

print(count)

# Problem - 03
# 👉 শুধু even number নিয়ে new list বানাও

arr = [1, 2, 3, 4, 5]

even = []
for i in range(len(arr)):
    if arr[i] %2 == 0:
        even.append(arr[i])
    
print(even)

# Problem - 04
# 20 এর বেশি number এর sum বের করো

arr = [10, 25, 30, 5, 40]

sum = 0

for i  in range(len(arr)):
    if arr[i] > 20:
        sum += arr[i]

print(sum)

# Probelm - 05
# minimum বের করো (condition use করে)

arr = [3, 8, 2, 9, 10]

min = arr[0]

for i in range(len(arr)):
    if min > arr[i]:
        min = arr[i]

print(min)
```

```python
# problem - 02
# 👉 odd number গুলো count করো

arr = [2, 3, 4, 5, 6]

count = 0
for i in range(len(arr)):
    if arr[i]%2 == 1:
        count += 1

print(count)
```
