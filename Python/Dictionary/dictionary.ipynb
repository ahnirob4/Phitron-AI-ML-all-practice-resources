### Python Dictionary

Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

    Important: As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

Dictionaries are changeable, meaning that we can change, add or remove items after the dictionary has been created.



```python
dict = {
    'Name' : 'Ashik',
    'Age'  : 20,
    'Address': 'Riyadh',
    'Number' : [1,2,3,4]
}

print(dict)
print(dict.get('Address'))
```

### New Value insert 
    note: If there is a duplicate key, it will always take the new key.

```python
dict = {
    'Name' : 'Ashik',
    'Age'  : 20,
    'Address': 'Riyadh',
    'Number' : [1,2,3,4]
}

dict['Name'] = 'Dipankar'
dict['Name'] = 'Sulaiman'
print(dict)
```

### Keys, values, get


keys - The view object contains the keys of the dictionary, as a list.

values - The view object contains the values of the dictionary, as a list.

get - Returns the value of the specified key

```python
dict = {
    'Name' : 'Ashik',
    'Age'  : 20,
    'Address': 'Riyadh',
    'Number' : [1,2,3,4]
}

keys = dict.keys()
# print(keys)

values = dict.values()
# print(values)

get_ = dict.get('Ambani', 0)
print(get_) 
```

### Update , Delete, Pop

update - Updates the dictionary with the specified key-value pairs

#### What is the difference between Python del and pop?

Del is a statement and pop is a method. thats the difference.

    Note: Pop returns a removed value, but del does not.



```python
dict = {
    'Name' : 'Ashik',
    'Age'  : 20,
    'Address': 'Riyadh',
    'Number' : [1,2,3,4]
}

dict.update({'Car' : 'Mercedes'})

print(dict)

# Delete

del dict['Number']
print(dict)

# Pop 

dict.pop('Name')
print(dict)
```

## Dictionary Loop

৩ ভাবে iterate — সব জানলে power আসে

```python
# Method 1: Keys এর উপর loop (Default)

student = {"name": "Rahim", "age": 20, "city": "Dhaka"}

for stu in student:
    print(stu)

# একই কাজ:

for key in student.keys():
    print(key)
```

```python
# Method 2: Values এর উপর loop

student = {"name": "Rahim", "age": 20, "city": "Dhaka"}

for values in student.values():
    print(values)
```

```python
# Method 3: .items() — Key এবং Value একসাথে

student = {"name": "Rahim", "age": 20, "city": "Dhaka"}

for keys, values in student.items():
    print(f'{keys}: {values}')
```

```python
#  Loop করার সময় Modify করা — বিপদ!

# d = {"a": 1, "b": 2, "c": 3}

# ❌ এটা করা যাবে না! loop চালানর সময় dictionary তে modify যায়না 
# for key in d:
#     if d[key] == 2:
#         del d[key] # RuntimeError
        
# print(d)

# Bypass 
# ✅ Safe way: d.keys() কে list() এ convert করে loop করা।

d = {"a": 1, "b": 2, "c": 3}

for key in list(d.keys()):
    if d[key] == 2:
        del d[key]
print(d)


```

## Dictionary Tricks & Advanced

```python
# Condition সহ comprehension
scores = {"Rahim": 85, "Karim": 45, "Jamal": 72, "Nadia": 91}

passed = {name: marks for name, marks in sorted(scores.items(), key= lambda x: x[1], reverse=True) if marks >= 60}

print(passed)
```

```python
# Invert — value কে key বানাও

scores = {"Rahim": 85, "Karim": 45, "Jamal": 72, "Nadia": 91}

invert = {value: key for key, value in sorted(scores.items(), key= lambda x: x[1])}
print(invert)
print(invert[91])
print(invert[72])
```

```python
from collections import Counter

words = ["apple", "banana", "apple", "cherry", "banana", "apple"]

count = Counter(words)
print(count)
print(count.most_common(2))
```

```python
from collections import defaultdict

word_count = defaultdict(int)
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]

for word in words:
    word_count[word] += 1

print(dict(word_count))

invert = {value: key for key, value in word_count.items()}
print(invert)


new = []
for key in invert:
    new.append(key) 
    avg = sum(new)/len(new)
print(new)
print(int(avg))

```

```python
# list default — missing key এ [] return করবে

from collections import defaultdict

group = defaultdict(list)

group['fruit'].append('apple')
group['fruit'].append('orange')
group['fruit'].append('appricot')
group['fruit'].remove('apple')

print(dict(group))
```
