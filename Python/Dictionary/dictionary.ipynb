### Python Dictionary

Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

    Important: As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

Dictionaries are changeable, meaning that we can change, add or remove items after the dictionary has been created.



```python
dict = {
    'Name' : 'Ashik',
    'Age'  : 20,
    'Address': 'Riyadh',
    'Number' : [1,2,3,4]
}

print(dict)
print(dict.get('Address'))
```

### New Value insert 
    note: If there is a duplicate key, it will always take the new key.

```python
dict = {
    'Name' : 'Ashik',
    'Age'  : 20,
    'Address': 'Riyadh',
    'Number' : [1,2,3,4]
}

dict['Name'] = 'Dipankar'
dict['Name'] = 'Sulaiman'
print(dict)
```

### Keys, values, get


keys - The view object contains the keys of the dictionary, as a list.

values - The view object contains the values of the dictionary, as a list.

get - Returns the value of the specified key

```python
dict = {
    'Name' : 'Ashik',
    'Age'  : 20,
    'Address': 'Riyadh',
    'Number' : [1,2,3,4]
}

keys = dict.keys()
# print(keys)

values = dict.values()
# print(values)

get_ = dict.get('Ambani', 0)
print(get_) 
```

### Update , Delete, Pop

update - Updates the dictionary with the specified key-value pairs

#### What is the difference between Python del and pop?

Del is a statement and pop is a method. thats the difference.

    Note: Pop returns a removed value, but del does not.



```python
dict = {
    'Name' : 'Ashik',
    'Age'  : 20,
    'Address': 'Riyadh',
    'Number' : [1,2,3,4]
}

dict.update({'Car' : 'Mercedes'})

print(dict)

# Delete

del dict['Number']
print(dict)

# Pop 

dict.pop('Name')
print(dict)
```

```python
# Method 4: .clear() — সব মুছো কিন্তু dict থাকুক
# del student আর student.clear() এর পার্থক্য:
#   • del student → ব্যাগটাই ফেলে দিলাম
#   • student.clear() → ব্যাগ রইলো, ভেতরের সব বের করলাম


subjects = {
    'Math': 38, 
    'Physics': 47, 
    'Chemistry': 89, 
    'Biology': 63,
}

subjects.clear()
print(subjects)
print(len(subjects))
```

## Dictionary Loop

৩ ভাবে iterate — সব জানলে power আসে

```python
# Method 1: Keys এর উপর loop (Default)

student = {"name": "Rahim", "age": 20, "city": "Dhaka"}

for stu in student:
    print(stu)

# একই কাজ:

for key in student.keys():
    print(key)
```

```python
# Method 2: Values এর উপর loop

student = {"name": "Rahim", "age": 20, "city": "Dhaka"}

for values in student.values():
    print(values)
```

```python
# Method 3: .items() — Key এবং Value একসাথে

student = {"name": "Rahim", "age": 20, "city": "Dhaka"}

for keys, values in student.items():
    print(f'{keys}: {values}')
```

```python
#  Loop করার সময় Modify করা — বিপদ!

# d = {"a": 1, "b": 2, "c": 3}

# ❌ এটা করা যাবে না! loop চালানর সময় dictionary তে modify যায়না 
# for key in d:
#     if d[key] == 2:
#         del d[key] # RuntimeError
        
# print(d)

# Bypass 
# ✅ Safe way: d.keys() কে list() এ convert করে loop করা।

d = {"a": 1, "b": 2, "c": 3}

for key in list(d.keys()):
    if d[key] == 2:
        del d[key]
print(d)


```
