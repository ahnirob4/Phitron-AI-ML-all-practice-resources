## Advance File Handling
### CSV = Comma Separated Values

👉 এটা আসলে একটা simple text file

👉 শুধু comma দিয়ে data আলাদা করা

CSV Reader line কে লিস্টে কনভার্ট করে

reader কি দেয়?

list of values

প্রতিটা row → list



```python
# CSV Read

import csv

with open('name.csv', 'r') as file:

    reader = csv.reader(file)
    for read in reader:
        print(read)
```

```python
# CSV Write
# newline="" কেন?

# 👉 না দিলে extra blank line আসতে পারে
# 👉 এটা Python official recommendation ✔️


with open('name.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    
    writer.writerow(['Name', 'Age', 'City', 'Marks'])
    writer.writerow(['Ashik', 33, 'Cumilla', 88])
    writer.writerow(['Ashraf', 65, 'Dhaka', 79])
    writer.writerow(['Osama', 43, 'Noakhali', 58])
    writer.writerow(['Nilufa', 23, 'Sylhet', 93])

with open('name.csv', 'r') as file:
    reader = csv.reader(file)

    for read in reader:
        print(read)
```

```python
# CSV Append
import csv
with open('name.csv', 'a', newline='') as file:

    writer = csv.writer(file)

    writer.writerow(['Shahadat', 28, 'Tangail', 58])

with open('name.csv', 'r') as file:

    reader = csv.reader(file)
    for read in reader:
        print(read)
```

```python
import csv

with open('name_2.csv', 'w', newline='') as file:

    fieldname = ['Name', 'Age','City', 'Marks']

    writer = csv.DictWriter(file, fieldnames= fieldname)

    writer.writeheader() 
    writer.writerow({'Name': 'Ashik','Age': 33,'City': 'Cumilla', 'Marks': 90})
    writer.writerow({'Name': 'Ashraf','Age': 65,'City': 'Dhaka', 'Marks': 78})
    writer.writerow({'Name': 'Osama','Age': 43,'City': 'Noakhali', 'Marks': 73})
    writer.writerow({'Name': 'Nilufa','Age': 22,'City': 'kushtia', 'Marks': 95})
    writer.writerow({'Name': 'Shahadat','Age': 45,'City': 'Barishal', 'Marks': 85})
    writer.writerow({'Name': 'Nazifa','Age': 18,'City': 'Nilfamari', 'Marks': 75})

with open('name_2.csv', 'r') as file:
    reader = csv.DictReader(file)
    for read in reader:
        print(read) 
```

```python
import csv

with open('people.csv', 'r') as file:

    reader = csv.DictReader(file)

    cleaned_row = []
    for read in reader:
        
        cleaned = {}
        for key,value in read.items():
            if value:
                cleaned[key] = value.strip()
            else:
                cleaned[key] = ''
        cleaned_row.append(cleaned)
        print(cleaned_row)
             
```
