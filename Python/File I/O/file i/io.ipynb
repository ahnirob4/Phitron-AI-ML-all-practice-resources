```python
# pointer effect

"""
file pointer 0 থেকে শুরু হয়। প্রাথমিক অবস্থায় 0 কিন্তু যখন read করে ফেলি তখন যতটুকু read করি তার শেষ বিন্দুতে গিয়ে অবস্থান করে।

Common Mistakes ❌
❌ read() multiple বার দিলে empty আসতে পারে
👉 কারণ pointer শেষে চলে যায়
❌ বড় file-এ read() ব্যবহার করা
👉 memory crash হতে পারে
❌ newline (\n) handle না করা
❌ file close না করা
"""

file = open('test.txt', 'r')

print(file.read(5)) # প্রথম ৫ টা index read করেছে। 
print(file.read(7).strip()) # পরবর্তী ৭ টা index read করেছে।
print(file.read(9).strip()) # পরবর্তী ৯ টা index read করেছে।
print(file.read().strip())  # এটা বাকী সব গুলো ইনডেক্স read করেছে। এখন আবার প্রিন্ট করলে pointer এর অবস্থান শেষে দেখাবে। 
```

### File Write

পুরোনো সব মুছে নতুন লিখা → w mode

```python
"""
write() কিভাবে কাজ করে?
আগের সমস্থ ডাটা মুছে দিবে নতুন করে লিখবে ।
শুধু string নেয়
যতটুকু লিখবে → return করে number of characters
number লিখতে গেলে error. এটাকে string এ কনভার্ট করে লিখতে হবে। 

👉 "w" mode = Danger Mode
👉 কি হবে?

যদি file থাকে → সব delete ❌
না থাকলে → নতুন create ✔️

"""

file = open('test.txt', 'w')

x = file.write('Python is the most popular language in the world.')
print(x)

file.close()
```

```python
"""
writelines() কিভাবে কাজ করে?
list/iterable নেয়
নিজে newline (\n) add করে না ⚠️

"""

file = open('test.txt', 'w')

x = file.writelines(['Hello\n', 'Python\n'])

print(x)
```

### File Pointer (tell & seek)


```python
"""
ধরো তুমি একটা বই পড়ছো 📖

তুমি এখন কোন লাইনে আছো → এটা হলো pointer position
তুমি চাইলে আবার শুরুতে যেতে পারো
বা মাঝখানে jump করতে পারো

👉 Python-এ file pointer ঠিক এই কাজটাই করে

Common Mistakes 
❌ read() করার পর আবার read করলে empty আসে
👉 কারণ pointer শেষে চলে গেছে
❌ seek() ভুল position দেওয়া
❌ মনে করা line অনুযায়ী move হয়
👉 আসলে byte অনুযায়ী
❌ binary vs text difference না বোঝা
"""

with open('position.txt', 'r') as file:

    print(file.tell())
    print(file.read(5))
    print(file.tell())
    print(file.read().strip())
    print(file.tell())
```

```python
# Custom File Reader
# Task:

# user input দিবে: start position
# user input দিবে: কত character পড়বে
# program সেই অনুযায়ী read করবে

with open('position.txt', 'r', encoding="utf-8") as file:

    start = int(input('Start your position: '))
    length = int(input('How many characters to read: '))

    file.seek(start)
    x = file.read(length)
    
    print(start)
    print(length)
    print(x)
    print(file.tell())
```

### Exception Handling (Error Handle)

```python
"""
    Create a text file named names.txt containing a list of names.
Some names may have extra spaces, inconsistent capitalization (like “aLiCe” or “ bob ”), or appear more than once.
Write a Python program to clean the data by:

1.  Removing extra spaces.
2.  Converting all names so that only the first letter is capitalized.
3.  Removing duplicate names. 
4.  Sorting the names alphabetically.
Finally, save the cleaned list into a new file named clean_names.txt.
"""

with open('names.txt', 'r') as file:
    
    cleaned_name = []

    for name in file:
        cleaned = name.strip().title()
        if cleaned not in cleaned_name:
            cleaned_name.append(cleaned)
    cleaned_name.sort()
    
with open('clean_names.txt', 'w') as file:
    for name in cleaned_name:
        file.write(name + '\n')
```

```python
# Problem of module 3

# 1. Count number of line
# 2. Count number of word
# 3. Count number of character
from functools import reduce
with open('sample.txt', 'r') as file:

# Count number of line

    read = file.readlines()
    total_line = len(read)
    print(read)
# Count number of word

    num = list(map(lambda x: len(x.split()), read))
    print(num)
    total_word = reduce(lambda x,y: x+y, num)
    print(total_word)

# Count number of character

    cleaned = list(map(lambda x:  len(x.strip().replace(' ','')), read))
    total_char = reduce(lambda x,y:  x+y, cleaned)

    print(total_char)

with open('string.txt', 'w') as file:

    file.write(f'Total number of line: {total_line}\nTotal number of word: {total_word}\nTotal number of character: {total_char}')
```
