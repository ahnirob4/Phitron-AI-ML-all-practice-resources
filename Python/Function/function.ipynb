### Parameter and Argument

Parameter কী?

👉 Function define করার সময় যেটা লিখি:

Argument কী?

👉 Function call করার সময় যেটা পাঠাই:

👉 Parameter = placeholder
👉 Argument = real value

```python
def greet(name): # name = parameter
    print(f'Say hi! {name}')

greet('Ashik') # Ashik = Argument
```

```python
def add(a,b):  # আমি চাইলে একাধিক parameter নিতে পারবো । 
    print(a+b)

add(5,10)
```

```python
def info(name, age, marks): # আমি কত গুলো Parameter নিয়েছি সেই অনুযায়ী ততগুল argument দিতে হবে নয়তো error আসবে।  
    print(f'Name: {name}, Age: {age}, Marks: {marks}')

info('Isra Bilzik', 38, 78.5)
```

```python
# Multiply

def multiply(a,b):
    print(a*b)

multiply(6, 9)
```

```python
def fullname(fname, lname):
    print(f'{fname} {lname}')

fullname('Abdullah al', 'Kabir')
```

```python
# Bill

def bill(price,tax):
    tax = (price * tax)/100
    print(price + tax)

bill(100,15)
```

```python
# Compare 

def compare(a,b):
    if a> b:
        print(f'A is greater then B')
    else:
        print(f'A is less then B')

compare(55,78)
```

### Print VS Return

print শুধু দেখায় কিন্তু return resuable value দেয়। 

print

👉 শুধু console-এ দেখায়
👉 function শেষ → value হারিয়ে যায় ❌

return

👉 value calculate করে
👉 caller-এর কাছে পাঠায়

```python
def add(a,b):
    print(a+b)

add(3,6)

def add(a,b):
    return a+b

result = add(3,6)
print(result)
```

## Positional Argument and Keyword

👉 Function call করার সময়:
👉 order (position) অনুযায়ী value assign হয়


```python
def info(name, marks):  # 1st position name , 2nd position marks
    return f'Name: {name}, Marks: {marks}'

x = info('Antique', 84)
print(x)
```

```python
# Generator

lst = [x for x in range(0,501)]

# for i in range(0, len(lst), 20):
#     print(lst[i:i+20])

def data_loader(chunk_size):
    for i in range(1, len(lst), chunk_size):
        yield lst[i: i+chunk_size]

x = data_loader(10)
# print(next(x))
# print(next(x))
# print(next(x))
# print(next(x))

for chunk in x:
    for num in chunk:
        if num > 1:
            for i in range(2, int(num**.5)+ 1):
                if (num % i == 0):
                    break
            else:
                print(num)
```

### Filter()

👉 filter() একটা ছাঁকনি (filter machine)

👉 এটা একটা condition (True/False function) নেয়

👉 তারপর iterable থেকে শুধু True হলে সেই element রাখে

👉 filter(function, iterable)

👉 এখানে function অবশ্যই True/False return করবে

👉 filter() list দেয় না

👉 এটা filter object (iterator) return করে

```python
# filter(function(return: True/False), iterable)

num = [x for x in range(1, 21)]

def odd(x):
    return x%2 == 1

result = list(filter(odd, num))
print(result)  
```

```python

names = ["Sonja", "Bradly", "Deshaun", "Karlie", "Miguelangel", "Shanice", "Daren", "Karlie", "Ricardo","Layne"]

result = list(filter(lambda x: len(x) > 6, names))

print(result)
```

```python
# Map + Filter
num = [x for x in range(1,21)]
result = list(map(lambda x: x**2, filter(lambda x: x%2 ==1, num)))

print(result)
```

```python
# আগে filter করো (>=10), তারপর square করো

numbers = [5, 10, 15, 20]

result = list(map(lambda x: x**2, filter(lambda x: x >= 10, numbers)))
print(result)
```

### Map , Filter, Lambda Function Combo Problem

```python
# problem 1 convert to string 


nums = [1, 2, 3, 4, 5]

result = list(map(lambda x: str(x), nums))
result

```

```python
# problem 2 filter() দিয়ে শুধু length > 5 এমন word রাখো

words = ["apple", "banana", "kiwi", "mango"]

result = list(filter(lambda x: len(x) > 5, words))
result
```

```python
# problem 4 filter() দিয়ে শুধু 3 দিয়ে divisible number রাখো

nums = [3, 5, 8, 11, 15,17]

result = list(filter(lambda x: x % 3==0, nums))

result
```

```python
# problem 5 odd numbers filter তারপর cube করো

nums = [1, 2, 3, 4, 5]

result = list(map(lambda x: x**3, filter(lambda x: x%2==1, nums)))

result
```

```python
# problem 6 10% discount apply করো (map())

prices = [100, 200, 300, 400]
# disc = prices * 10/100

discount = list(map(lambda x: x - (x*10/100), prices ))
discount
```

```python
# problem 7  👉 string → int convert করো 👉 তারপর even number রাখো

data = ["1", "2", "3", "4", "5"]

convert = list(filter(lambda x: x%2==0, map(lambda x: int(x), data)))

convert
```

```python
# problem 8 
# marks >= 60 filter
# 👉 তারপর grade বানাও:
# =80 → "A"
# =60 → "B"

students = [50, 80, 30, 90, 60]

filt = list(filter(lambda x: x >= 60, students))
print(filt)
result = ['Grade: A' if x >= 80 else 'Grade: B' for x in filt]
print(result)
new = []
for i in filt:
    if i >= 80:
        new.append('Grade: A')
    else:
        new.append('Grade: B')
    
print(new)
```
