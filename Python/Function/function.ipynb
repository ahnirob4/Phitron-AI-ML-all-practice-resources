### Parameter and Argument

Parameter কী?

👉 Function define করার সময় যেটা লিখি:

Argument কী?

👉 Function call করার সময় যেটা পাঠাই:

👉 Parameter = placeholder
👉 Argument = real value

```python
def greet(name): # name = parameter
    print(f'Say hi! {name}')

greet('Ashik') # Ashik = Argument
```

```python
def add(a,b):  # আমি চাইলে একাধিক parameter নিতে পারবো । 
    print(a+b)

add(5,10)
```

```python
def info(name, age, marks): # আমি কত গুলো Parameter নিয়েছি সেই অনুযায়ী ততগুল argument দিতে হবে নয়তো error আসবে।  
    print(f'Name: {name}, Age: {age}, Marks: {marks}')

info('Isra Bilzik', 38, 78.5)
```

```python
# Multiply

def multiply(a,b):
    print(a*b)

multiply(6, 9)
```

```python
def fullname(fname, lname):
    print(f'{fname} {lname}')

fullname('Abdullah al', 'Kabir')
```

```python
# Bill

def bill(price,tax):
    tax = (price * tax)/100
    print(price + tax)

bill(100,15)
```

```python
# Compare 

def compare(a,b):
    if a> b:
        print(f'A is greater then B')
    else:
        print(f'A is less then B')

compare(55,78)
```

### Print VS Return

print শুধু দেখায় কিন্তু return resuable value দেয়। 

print

👉 শুধু console-এ দেখায়
👉 function শেষ → value হারিয়ে যায় ❌

return

👉 value calculate করে
👉 caller-এর কাছে পাঠায়

```python
def add(a,b):
    print(a+b)

add(3,6)

def add(a,b):
    return a+b

result = add(3,6)
print(result)
```

```python
def add(a,b):
    print(a+b)

result = add(12,13)
print(result)

# 😱 কেন None?

# 👉 কারণ function কিছু return করে নাই
# 👉 default return = None
```

```python
def add(a,b):
    return a+b

x = add(12,13)
y = add(7,8)
z = add(15,16)


print(x+y+z)

# ⚡ এটা শুধু print দিয়ে impossible 
```

## Positional Argument and Keyword

👉 Function call করার সময়:
👉 order (position) অনুযায়ী value assign হয়


```python
def info(name, marks):  # 1st position name , 2nd position marks
    return f'Name: {name}, Marks: {marks}'

x = info('Antique', 84)
print(x)
```

```python
# keyword argument

def info(name, age):
    print(f'{name}, {age} years')
    
info(age=35, name= 'Renesa')
```

## Iterator and Generator

```python
# যেই ডাটা টাইপ গুলোকে আমরা অ্যাক্সেস করতে পারিনা সেগুলোকে আমরা iterator দিয়ে অ্যাক্সেস করতে পারি 
# iterator শুধু ইনডেক্স এর লেন্থ পর্যন্ত চালাতে পারবো এর পরে error দিবে ।

new_set = {10,20,30,40,50}

s = iter(new_set)
print(next(s))
print(next(s))
print(next(s))
print(next(s))
print(next(s))

for i in new_set:
    print(i)
```

```python
dict = {
    'Name' : 'Ashik',
    'Age'  : 20,
    'Address': 'Riyadh',
    'Number' : [1,2,3,4]
}

s = iter(dict)
print(type(s))
print(next(s))
print(next(s))
print(next(s))
print(next(s))
```

```python
# Generator

lst = [x for x in range(0,501)]

# for i in range(0, len(lst), 20):
#     print(lst[i:i+20])

def data_loader(chunk_size):
    for i in range(1, len(lst), chunk_size):
        yield lst[i: i+chunk_size]

x = data_loader(10)
# print(next(x))
# print(next(x))
# print(next(x))
# print(next(x))

for chunk in x:
    for num in chunk:
        if num > 1:
            for i in range(2, int(num**.5)+ 1):
                if (num % i == 0):
                    break
            else:
                print(num)
```

### Map()

```python
# map(function, iterable)

number = [2,4,6,8]

def double(x):
    return x**2

x = map(double, number)

print(list(x))
```
