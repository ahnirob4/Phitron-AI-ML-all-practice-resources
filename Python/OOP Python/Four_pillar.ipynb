```python
# PasswordManager

# private password
# method:
# set_password(new_pass)
# check_password(input_pass)

# 👉 condition:

# direct access করা যাবে না

class PasswordManager:
    def __init__(self,new_pass):
        self.__new_pass = new_pass

    def set_password(self, input_pass):
        self.__new_pass = input_pass
    
    def check_password(self, input_pass):
        if self.__new_pass == input_pass:
            return f'Your password is Correct. Your pass: {self.__new_pass}'
        
        else:
            return f'Your password is incorrect.'
        
reg = PasswordManager('abc123')

reg.set_password('1243')
x = reg.check_password('1243')
print(reg._PasswordManager__new_pass)
```

```python
# Problem: SecureBankAccount

# 👉 Requirements:

# private __balance
# method:
# deposit(amount)
# withdraw(amount)
# get_balance()
# ❗ Conditions:
# withdraw → check (balance >= amount)
# negative deposit allow না
# return use করবে (print না)
# direct access করা যাবে না

class SecureBankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        if amount <= 0:
            return 'Deposit amount must be positive'
        self.__balance += amount
        return self.__balance

    def withdraw(self, amount):
        if amount <= 0 :
            return 'You can not give any negative number.'

        if self.__balance >= amount:
            self.__balance -= amount
            return self.__balance
        else:
            return 'You do not have sufficient funds. Thank you!'
        
    def get_balance(self):
        return self.__balance
    

my_balance = SecureBankAccount(5000)
print(my_balance.deposit(3000))
print(my_balance.withdraw(1000))
```

### Abstraction

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class dog(Animal):
    def sound(self):
        return 'Woof'
    
class cat(Animal):
    def sound(self):
        return 'Meow'
    


d = dog()
e = cat()
print(d.sound())
print(e.sound())
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
        
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.1416 * self.radius * self.radius
    

r = Rectangle(3,5)
c = Circle(5)

print(r.area())
print(c.area())
```

```python
from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self):
        pass

class CreditCard(Payment):

    def pay(self, amount):
        return f'Paid {amount} using CreditCard.'
    
class PayPal(Payment):
    def pay(self, amount):
        return f'Paid {amount} using PayPal.'
    

c = CreditCard()
p = PayPal()
print(c.pay(500))
print(p.pay(500))
```

```python
# Requirements:
# abstract class: Vehicle
# method: start()
# Child class:
# Car → "Car started"
# Bike → "Bike started"
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        return 'Started...'

class Car(Vehicle):
    def start(self):
        return 'Car started...'
    
class Bike(Vehicle):
    def start(self):
        return "Bike started..."
    
c = Car()
b = Bike()

print(c.start())
print(b.start())
```

```python
# 👉 Problem:

# Account

# deposit()

# SavingsAccount

# inherit + interest()

class Account:
    def __init__(self, balance = 0, rate = 0 , time = 0):
        self.balance = balance
        self.rate = rate
        self.time = time

    def deposit(self, amount):
        self.balance += amount


class SavingsAccount(Account):
    def __init__(self, balance=0, rate=0, time=0):
        super().__init__(balance, rate, time)

    def interest(self):
        interest_amount = (self.balance * ((1+(self.rate)/100) ** self.time))
        return interest_amount
    def get_balance(self):
        return self.balance
    

s = SavingsAccount(20000,5,1)  # amount = 20000, rate = 5% , time = 5 years
print(s.get_balance())
print(s.interest())

```

```python
class InterestCalculator:
    def __init__(self, principal, rate, time):
        self.principal = principal
        self.rate = rate
        self.time = time

    def simple_interest(self):
        return (self.principal * self.rate * self.time) / 100

    def compound_interest(self):
        return self.principal * ((1 + self.rate/100) ** self.time) - self.principal

# ব্যবহার
calc = InterestCalculator(1000, 5, 2)
print("সরল সুদ:", calc.simple_interest())
print("চক্রবৃদ্ধি সুদ:", calc.compound_interest())   
```
