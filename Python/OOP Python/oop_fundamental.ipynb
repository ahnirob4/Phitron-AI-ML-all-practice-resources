```python
# 👉 Problem:

# BankAccount class বানাও

# attribute: owner, balance
# method:
# deposit(amount)
# withdraw(amount)

# 👉 Condition:

# withdraw করলে balance কমবে
# print final balance

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f'Balance after deposit: {self.balance}')


    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f'Balance after withdraw: {self.balance}') 
        else:
            print('Insufficient Balance')   

bank = BankAccount('Aslam', 5000)

bank.deposit(200)
bank.withdraw(500)
bank.withdraw(4700)
```

```python
class ATM:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f'Balance after deposit: {self.balance}')


    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f'Balance after withdraw: {self.balance}') 
        else:
            print('Not enough money')   

bank = ATM(5000)

bank.deposit(200)
bank.withdraw(500)
bank.withdraw(4700)
```

```python
# Product class

# class variable: discount = 10%
# instance: name, price

# 👉 method:

# final_price() → discount apply করে return করবে

class Product:
    discount = 10/100

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def final_price(self ):
        return self.price - (self.price * self.discount)
    
customer1 = Product('Anik', 350)

print(customer1.name)
print(customer1.price)
print(customer1.final_price())
print(customer1.discount)     
```

```python
# 🟢 Easy (3টা)
# Person → name, method: introduce()
# Car → brand, method: show_brand()
# Dog → name, method: speak()

# 🟡 Medium (2টা)
# Student
# name, marks
# method: is_pass() → marks >= 40 হলে pass
# Circle
# radius
# method: area()

# problem -01
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        return self.name
    
# problem - 02

class Car:
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        return self.brand
    
# problem - 03

class Dog:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return self.name
    

# problem - 04

class Student:
    def __init__(self, name, marks):
        self.name = name 
        self.marks = marks

    def is_pass(self):
        if self.marks >= 40:
            return f'Name: {self.name} Result: {self.marks} Pass'
        else:
            return 'you are fail'
        
stu1 = Student('Jotindronath', 50)
print(stu1.is_pass())

# problem - 05

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.1416 * self.radius * self.radius
    
area1 = Circle(5)
print(area1.area())
```

```python
# Counter class

# attribute: count = 0
# method:
# increment() → count +1
# decrement() → count -1

# 👉 multiple object create করে check করো:

# each object কি আলাদা count রাখছে?

class Counter:
    def __init__(self,count=0):
        self.count = count

    def increment(self):
        self.count += 1
        return self.count
    
    def decrement(self):
        self.count -= 1
        return self.count
    
count1 = Counter(5)
print(count1.increment())
print(count1.decrement())
```
