# List as a Stack

```python
## Adding new element inside the stack

stack = []

stack.append(50)    ## add first elemnt [50]
stack.append(150)   ## add second elemnt [150]
stack.append(20)    ## add third elemnt [20]
stack.append(70)    ## add fourth elemnt [50]
stack.append(10)    ## add fifth elemnt [10]
stack.append(90)    ## add sixth elemnt [90]
stack.sort()
print(stack)

stack.pop(0)  ## remove last element [150]
stack.pop(0)  ## remove last element [90]
stack.pop(0)  ## remove last element [70]
stack.pop(0)  ## remove last element [50]
print(stack)
```

# List as a Queue

```python
## Enqueue and Dequeue
## First in first out - FIFO

from collections import deque
queue = deque()

queue.append(10)
queue.append(20)
queue.append(30)
queue.append(40)
queue.append(50)

print(queue)

queue.popleft()   ## dequeue first element [10]
queue.popleft()   ## dequeue first element [20]
queue.popleft()   ## dequeue first element [30]

print(queue)

## Check empty

print(len(queue) == 0)

## Queue Peek

print(queue[0])   ## Peek
```

# Practice Problem

Given a number N and an array A of N numbers. Print the lowest number and its position.

Note: if there are more than one answer print first one's position.

Input
First line contains a number N (2 ≤ N ≤ 1000) number of elements.

Second line contains N numbers (-105  ≤  Ai  ≤  105).

Output
Print the lowest number and its position (1-index).

```python
num = input()

arr = list(map(int, num.split()))

min_num = arr[0]
position = 1

for i in range(len(arr)):
    if arr[i] < min_num:
        min_num = arr[i]
        position = i + 1

print(arr)
print(min_num, position)
```

Given a number N and an array A of N numbers. Print the array after doing the following operations:

Find minimum number in these numbers.

Find maximum number in these numbers.

Swap minimum number with maximum number.

Input
First line contains a number N (2 ≤ N ≤ 1000) number of elements.

Second line contains N numbers ( - 105 ≤ Ai ≤ 105)

It's guaranteed that all numbers are distinct.

Output
Print the array after the replacement operation.

```python
n = int(input())

arr = list(map(int, input().split()))

max_index = 0
min_index = 0
for i in range(n):
    if arr[i] > arr[max_index]:
        max_index = i
    if arr[i] < arr[min_index]:
        min_index = i
    
arr[max_index], arr[min_index] = arr[min_index], arr[max_index]
print(*arr)
```

Given a number N and an array A of N digits (not separated by space). Print the summation of these digits.

Input
First line contains a number N (1  ≤  N  ≤  106) number of digits.

Second line contains N digits (0 ≤ Ai ≤ 9).

```python
n = int(input())

num = list(map(int, str(input())))
sum = 0

for i in range(n):
    sum += num[i] 

print(sum)
```
