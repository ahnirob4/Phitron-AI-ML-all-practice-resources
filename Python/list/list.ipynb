##### Python list

```python
list_int = [10,30,20,50,70,40,60] # list inside integer
list_flo = [10.56,30.45,20.78,50.18] # list inside Float
list_str = ['Apple', 'Orange', 'Lemon', 'Pineapple'] # list inside String
list_mix = [10,'Guava', 38.20, True] # list inside Mix value

## accessing list

print(list_int[3])  # 50
print(list_int[-1]) # 60 last index


## Slicing

print(list_flo[1:4]) # [30.45, 20.78, 50.18]
print(list_flo[:])   # [10.56, 30.45, 20.78, 50.18] print whole list
```

```python
## Adding new element
list_flo = [10.56,30.45,20.78,50.18] # list inside Float

list_flo.append(80)
list_flo.append(90)
list_flo.append(100)

print(list_flo)
```

```python
## Deleting existing element

list_str = ['Apple', 'Orange', 'Lemon', 'Pineapple'] # list inside String

list_str.pop() ## delete last item

print(list_str)
```

```python
## Deleting Specific Element

list_str = ['Apple', 'Orange', 'Lemon', 'Pineapple'] # list inside String

list_str.remove('Lemon') ## Input only the item name, and then that item will be removed from the list.
print(list_str)
```

```python
## Updating existing Element

list_mix = [10,'Guava', 38.20, True] # list inside Mix value

list_mix[1] = 'Pomegranate' ## Update specific element

print(list_mix)
```

```python
## List sort and reverse
list_int = [10,30,20,50,70,40,60] # list inside integer

list_int.sort()
print(list_int)

list_int.reverse() ## list reverse
print(list_int)
```

# List As a Stack

```python
## Adding new element inside the stack

stack = []

stack.append(50)    ## add first elemnt [50]
stack.append(150)   ## add second elemnt [150]
stack.append(20)    ## add third elemnt [20]
stack.append(70)    ## add fourth elemnt [50]
stack.append(10)    ## add fifth elemnt [10]
stack.append(90)    ## add sixth elemnt [90]
stack.sort()
print(stack)

stack.pop(0)  ## remove last element [150]
stack.pop(0)  ## remove last element [90]
stack.pop(0)  ## remove last element [70]
stack.pop(0)  ## remove last element [50]
print(stack)
```
