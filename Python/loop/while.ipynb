### While loop

👉 while loop হলো এমন একটা loop

👉 যা একটা condition True থাকা পর্যন্ত বারবার run করে

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
while True:
    user = input("Enter 'Exit' to stop: ")

    if user.lower() == 'exit':
        break
    else:
        print(user)
```

```python
password = ''
while password != '1234':
    password = input('Enter your password: ')

print('Access Granted')
```
