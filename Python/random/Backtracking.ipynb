## Backtracking

ধরো তুমি একটা maze (গোলকধাঁধা) এর ভিতরে আছো।

👉 তুমি এক রাস্তা ধরে এগোচ্ছো

👉 যদি দেখো রাস্তা বন্ধ ❌

👉 তাহলে তুমি আবার আগের জায়গায় ফিরে গিয়ে অন্য রাস্তা চেষ্টা করো

এই কাজটাই হলো Backtracking।

✔ সামনে যাও

✔ ভুল হলে পিছনে আসো

✔ অন্য পথ চেষ্টা করো

```python
def backtrack(path):
    
    #base case
    if len(path) == 3 :
        print(path)
        return

    for i in range(1,4):

        #choose
        path.append(i)

        #explore
        backtrack(path)

        #unchoose backtrack
        path.pop()
        
backtrack([])
```

```python
def backtrack(path):
     
     if len(path) == 2:
          print(path)
          return
     
     for i in [1,2]:
          path.append(i)
          backtrack(path)
          path.pop()

backtrack([])
```
