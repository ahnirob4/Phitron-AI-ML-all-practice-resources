## Recursion
⚠️ VERY IMPORTANT

👉 Recursion = 2 parts

Base case (stop)

Recursive call

```python
def numCount(n):
    if n==11:
        return
    print(n)
    numCount(n+1)

numCount(1)
```

```python
def repeatCount(n):
    if n == 0:
        return
    print('Hi!')
    repeatCount(n-1)
repeatCount(3)
```

```python
def sum_n(n):
    if n==0:
        return 0
    return n + sum_n(n-1)

result = sum_n(5)
print(result)
```

```python
def number_n(n):
    if n == 0:
        return 0
    
    print(n)
    number_n(n-1)

number_n(5)
```

### n এর factorial বের করো


```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    
    return n * fact(n-1)

result = fact(5)
print(result)
```

### Factorial 

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)

result = fact(4)
print(result)
```

```python
def inp_fact(n):
    if n == 0:
        return 1
    
    return n * inp_fact(n-1)

s = int(input('Enter your number: '))
result = inp_fact(s)
print(result)
```

## Recursion Call Stack 

👉 যখন function নিজেকে call করে:

প্রতিটা call → stack এ জমা হয়

শেষের call → আগে execute হয়

```python
def call_stack(n):
    if n == 0:
        return
    
    print('Calling: ', n)
    call_stack(n-1)
    print('Returning: ', n)

call_stack(5)
```

```python
def sum(n):
    if n == 0:
        return 0
    
    return n + sum(n-1)

result = sum(5)
print(result)
```
