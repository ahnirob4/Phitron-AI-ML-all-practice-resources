### Union & Intersection

```python

a = {1,2,3,5,7}
b = {3,4,5,6}

union = a | b
intersection = a & b

print(union)
print(intersection)
```

### Disjoint, Superset and Subset

```python
a = {1,2,3,4}
b = {1,2,3,4,5,6,7,8}
c = {5,6,7,8}

disjoint = a.isdisjoint(c)
superset = b >= a   # a.issuperset(b) 
subset = a <= b # a.issubset(b)

print(disjoint)
print(superset)
print(subset)
```

### Difference() and Symmetric_difference()

```python
set_1 = {1,2,3,4,5}
set_2 = {4,5,6,7}

defference = set_1 - set_2  # set_1.difference(set_2)
symmetric_defference = set_1 ^ set_2  # set_1.symmetric_difference(set_2)
print(defference)
print(symmetric_defference)
```
