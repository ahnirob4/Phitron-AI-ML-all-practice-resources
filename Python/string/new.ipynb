Given a number N. Print the digits of that number from right to left separated by space.

Input
First line contains a number T (1 ≤ T ≤ 10) number of test cases.

Next T lines will contain a number N (0 ≤ N ≤ 109)

```python
t = int(input())

for _ in range(t):
    N = input()
    
    for i in reversed(N):
        print(i, end=' ')
    print()
```

### Even, Odd, Positive and Negative

```python
n = int(input())
number = input().split()

even = 0
odd = 0
positive = 0
negative = 0

for i in range(n):
    x = int(number[i])
## EVEN OR ODD
    if x % 2 == 0:
        even += 1
    else:
        odd += 1

## POSITIVE OR NEGATIVE
    if x > 0:
        positive += 1
    if x < 0:
        negative += 1

print('Even:',even)
print('Odd:',odd)
print('Positive:',positive)
print('Negative:',negative)

```

### DIGITS

```python
t = int(input())

for i in range(t):
    num = int(input())
    if num == 0:
        print(0)
        continue
    while num > 0:
        digit = num % 10
        print(digit, end=' ')
        num = num //10
    print()
```
