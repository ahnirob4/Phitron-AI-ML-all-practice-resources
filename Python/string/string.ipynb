### Ascii Code

32 -> Space

(48-57) -> (0-9)

(65-90) -> (A-Z)

(97-122) -> (a-z)

(33–47, 58–64, 91–96, 123–126) -> Special character

Word to number -> ord()

Number to Word -> chr()

```python
asc = [97, 115, 108, 97, 109] # [65,83,76,65,77] 
for char in asc:
    # print(f'{char} -> {chr(char)}')
    print(chr(char), end="")

asc2 = 'aslam'
asc_small = []
for char in asc2:
    asc_small.append(ord(char))
    # print(ord(char))
print(f'\n{asc_small}')
```
